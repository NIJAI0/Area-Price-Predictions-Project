```python
import matplotlib.pyplot as plt
plt.scatter(x,y, color = "red", s=50)
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```
